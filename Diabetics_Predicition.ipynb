# Predicting Whether a Patient Has Diabetes or Not

### Importing the Relevant Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
#Splitting the dataset into independent features (x) and the dependent target (y)
x= data_set.iloc[:,:-1]
y= data_set.iloc[:,-1]
```

```python
print(x)
```

### Spliting the dataset into Traing Set and Test Set

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, stratify= y, random_state= 2)
```

### Appling the Feature Scaling for the dataset

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
print(x_train)
```

```python
print(x_test)
```

### Training SVM Model 

```python
model= svm.SVC(kernel= 'linear')
model.fit(x_train, y_train)
```

```python
# Accuaracy Score on Train Data
x_train_predict= model.predict(x_train)
AccuracyScore_Train= accuracy_score(x_train_predict, y_train)
print(AccuracyScore_Train) 
```

```python
#Finding the Accuary Score
y_predict=model.predict(x_test) 
print(y_predict)
```

```python
print(y_test)
```

```python
# Accuaracy Score on Test Data
AccuracyScore_Test= accuracy_score(y_test, y_predict)
print(AccuracyScore_Test) 
```

## Predicing for the Dynamic Input

```python
print("\nPlease Enter the following details: *** Note: Only enter numerical values ***\n")

input_dy1 = float(input("Pregnancies (whole number): "))
input_dy2 = float(input("Glucose: "))
input_dy3 = float(input("Blood Pressure: "))
input_dy4 = float(input("Skin Thickness: "))
input_dy5 = float(input("Insulin: "))
input_dy6 = float(input("BMI (e.g., 25.5): "))
input_dy7 = float(input("Diabetes Pedigree Function (e.g., 0.627): "))
input_dy8 = float(input("Age (whole number): "))

# Prepare input as a NumPy array
input1 = np.array([[input_dy1, input_dy2, input_dy3, input_dy4, 
                     input_dy5, input_dy6, input_dy7, input_dy8]])

# Standardize the input data
input_standardized = sc.transform(input1)

# Make prediction
Sol = model.predict(input_standardized)

# Display result
if Sol == [1]:
    print("\n You are diabetic. Please consult a doctor for proper medication and guidance.")
    print("âš¡ Tips for managing diabetes:")
    print("   - Maintain a balanced diet with low sugar and carbs.")
    print("   - Exercise regularly to keep your blood sugar levels stable.")
    print("   - Monitor glucose levels and follow medical advice.")
    print("   - Manage stress and ensure proper sleep.")
else:
    print("\nYou are not diabetic. Keep following a healthy lifestyle to stay safe!")
```
